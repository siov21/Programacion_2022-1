```python
import json
import requests
from pprint import pprint
response = requests.get("https://fp.josedomingo.org/lmgs/u05/ej2.json")
pendientes = json.loads(response.text)
```

```python
#1.¿Cuántas pruebas de idiomas están descritas en el documento?
print(len(pendientes))
```

```python
#2.Devuelve el título de las pruebas de nivel que van a durar más de dos horas.
horas=[]
for i in pendientes:
  if i["Horas"]>2:
    horas.append(i["Titulo"])
print('Los titulos que duran mas de 2 horas son:')
pprint(horas)
```

```python
from requests.models import encode_multipart_formdata
#3.De las pruebas de tipo “No Presencial” devuelve la URL de información.
nopre=[]
for i in pendientes:
  if i['TipoFormacion']=='No Presencial'or i['TipoFormacion']=='NoPresencial':
    nopre.append(i['URL'])
print('Las URL de las pruebas no presenciales son:')    
pprint(nopre)
```

```python
#4.Recibe el código de la prueba “ID” y muestra su título y profesores.
print('Pruebas Disponibles: ')
for i in pendientes:
  print(i['$id'])
id=(input('Ingrese el ID de la prueba del cual quiere ver la información: '))
for i in pendientes:
  if i['$id']==id:
    print('Ttulo:',i['Titulo'],'\nProfesores:')
    pprint(i['Profesorado'])
```
